## part two

```python
household_fillna = household.fillna(0,inplace=False)
household_process = household_fillna.drop(index = household_fillna.loc[household_fillna['idcluster'] == 0].index)
HH_satis = household_process[['idcluster','S_satisfloor','S_satishouse','S_satislife','S_cesds','S_pss']]
HH_satis
```

```python
# the average level of satisfaction feature across the cluster
HH_satis_mean = HH_satis.mean(axis=0)
HH_satis_mean
```

```python
# regression for cement floor in different rooms
def lr_cement(dependent):
    mod = smf.ols(formula='dependent ~ C(S_cementfloorkit) + C(S_cementfloordin) + C(S_cementfloorbat) + \
                        C(S_cementfloorbed)', data=household_process)
    res = mod.fit(cov_type='cluster',cov_kwds={'groups':household_process['idcluster']})
    variables = res.params.index
    coefficients = res.params.values
    p_values = res.pvalues
    standard_errors = res.bse.values
    l1, l2, l3, l4 = coefficients[1:], variables[1:], standard_errors[1:], p_values[1:]
    
    return l1,l3
```

```python
# regression for housing improvement
def lr_improvement(dependent):
    mod = smf.ols(formula='dependent ~ C(S_instcement) + C(S_instsanita) + C(S_restsanita) + \
                        C(S_constceili) + C(S_restowalls)+ C(S_improveany)', data=household_process)
    res = mod.fit(cov_type='cluster',cov_kwds={'groups':household_process['idcluster']})
    variables = res.params.index
    coefficients = res.params.values
    p_values = res.pvalues
    standard_errors = res.bse.values
    l1, l2, l3, l4 = coefficients[1:], variables[1:], standard_errors[1:], p_values[1:]
    
    return l1,l3
```

```python
index = ['Satisfaction with floor quality','Satisfaction with house quality','Satisfaction with quality of life',\
         'Depression scale (CES-D scale)','Perceived stress scale (PSS)']
cl_cement = ['Cement floor in kitchen','Cement floor in dining room','Cement floor in bathroom','Cement floor in bedroom']
cl_improvement = ['Installation of cement floor','Construction/expansion of sanitation facilities',\
                  'Restoration of sanitation facilities','Construction of ceiling',\
                  'Restoration of walls','Any house expansion (excluding installation of cement floors)']
table_cm = pd.DataFrame(index = index, columns = cl_cement)
table_im = pd.DataFrame(index = index, columns = cl_improvement)
```

```python
for i in range(len(index)):
        table_cm.iloc[i,:],_ = lr_cement(HH_satis.iloc[:,i+1])
        table_im.iloc[i,:],_ = lr_improvement(HH_satis.iloc[:,i+1])
```

```python
table_cm
```

```python
table_im
```

effect can be 100*coefficient/ mean

```python
effect_cm = pd.DataFrame(index = index, columns = cl_cement)
effect_im = pd.DataFrame(index = index, columns = cl_improvement)
z_cm = []
z_im = []

for i in range(len(index)):
        effect_cm.iloc[i,:] = table_cm.iloc[i,:]*100/HH_satis_mean[i+1]
        effect_im.iloc[i,:] = table_im.iloc[i,:]*100/HH_satis_mean[i+1]
        z_cm.append(np.array(effect_cm.iloc[i,:]))
        z_im.append(np.array(effect_im.iloc[i,:]))
```

```python
effect_cm
```

```python
effect_im
```

```python
def decimal_3(data):
    for i in range(data.shape[0]):
        for j in range(data.shape[1]):
            data[i][j] = round(data[i][j],3)
            
    return data
```

```python
z_cm = decimal_3(np.array(z_cm))
z_im = decimal_3(np.array(z_im))
```

```python
# visualization of cement floor (annotated heatmap)
import plotly.figure_factory as ff

fig = ff.create_annotated_heatmap(z_cm, x=cl_cement, y=index, annotation_text=z_cm)
fig.update_layout(
        title_text = 'Effect of the presence of cement floor in different rooms on satisfaction and stress indicators<br>',
        showlegend = True)
fig.write_html('2_cement_figure.html', auto_open=True)
```

```python
# visualization of housing improvement (annotated heatmap)

fig = ff.create_annotated_heatmap(z_im, x=cl_improvement, y=index, annotation_text=z_im)
fig.update_layout(
        title_text = 'Effect of the presence of housing improvement on satisfaction and stress indicators<br>',
        showlegend = True)
fig.write_html('2_improvement_figure.html', auto_open=True)
```
